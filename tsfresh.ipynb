```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
!pip install tsfresh
```

```python
from tsfresh import extract_features
```

```python
# Load the dataset
df = pd.read_csv("/content/Malaysia_final.csv")
# df.index = pd.to_datetime(df['Date'])
# df = df.drop(['Date'], axis='columns')
df = df.replace('#DIV/0!', np.nan)
df = df.dropna()
df
```

```python
from tsfresh.utilities.dataframe_functions import impute
# tsfresh feature extraction
extracted_features = extract_features(df, column_id='Date', column_sort='Date')
# Imputing the missing values
df_tsfresh = impute(extracted_features)

# Normalizing the features
scaler = MinMaxScaler(feature_range=(-1, 1))
X_tsfresh = scaler.fit_transform(df_tsfresh)

df_tsfresh.to_csv('tsfresh.csv', index=False)
```
